```python
# Convertir la columna a string para operaciones seguras
dataset['AHT'] = dataset['AHT'].astype(str)

# 1. Eliminar comas (separadores de miles)
# 2. Reemplazar '.' solitarios por NaN
# 3. Convertir a numérico
dataset['AHT'] = (
    dataset['AHT']
    .str.replace(',', '', regex=False)  # Eliminar comas
    .replace(r'^\.$', np.nan, regex=True)  # Reemplazar '.' exacto por NaN
    .replace('nan', np.nan)  # Manejar NaNs convertidos a string
)

# Conversión final a float64
dataset['AHT'] = pd.to_numeric(dataset['AHT'], errors='coerce')
```

```python
# Calcular el índice de eficiencia operativa
dataset["Efficiency Index"] = dataset["Handled Contacts"] / dataset["# of Experts"] / dataset["AHT"]

# Agrupar por semana
weekly_efficiency = dataset.groupby("Calendar Week")["Efficiency Index"].mean().reset_index()

weekly_efficiency
```

```python
# Convertir la columna de fecha y limpiar posibles valores no válidos
dataset['Calendar Week'] = pd.to_datetime(dataset['Calendar Week'], errors='coerce')
dataset = dataset.dropna(subset=['Calendar Week'])

# Calcular el índice de eficiencia
dataset['Operational Efficiency Index'] = (dataset['Handled Contacts'] /
                                           (dataset['# of Experts']) * dataset['AHT'])

# Agrupar por semana y calcular promedio
weekly_efficiency = dataset.groupby('Calendar Week')['Operational Efficiency Index'].mean().reset_index()

# Graficar
plt.figure(figsize=(12, 6))
plt.plot(weekly_efficiency['Calendar Week'], weekly_efficiency['Operational Efficiency Index'], marker='o')
plt.title('Índice de Eficiencia Operativa Semanal')
plt.xlabel('Semana (Fecha)')
plt.ylabel('Índice de Eficiencia Operativa')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Columnas necesarias
required_columns = [
    "Handled Contacts", "# of Experts", "AHT", "tNPS", "Issue Resolution (tNPS)"
]

# Asegurar que las columnas estén en formato numérico
for col in required_columns:
    dataset[col] = pd.to_numeric(dataset[col], errors='coerce')

# Eliminar filas con valores nulos en columnas clave
dataset = dataset.dropna(subset=required_columns)

# Índice de Eficiencia Centrada en el Cliente (IECC)
dataset['IECC'] = (dataset['tNPS'] * dataset['Issue Resolution (tNPS)']) / dataset['# of Experts']

# Eficiencia Total (peso igual para ambos)
dataset['Total Efficiency Index'] = 0.5 * dataset['IET'] + 0.5 * dataset['IECC']

# Agrupar por semana
weekly_efficiency = dataset.groupby('Calendar Week')[['IET', 'IECC', 'Total Efficiency Index']].mean().reset_index()
```

```python
# Rehacer la gráfica con los nombres de columna correctos
plt.figure(figsize=(12, 6))
plt.plot(weekly_efficiency['Calendar Week'], weekly_efficiency['IECC'], marker='o', label='IECC (Eficiencia centrada en el cliente)')
plt.plot(weekly_efficiency['Calendar Week'], weekly_efficiency['Total Efficiency Index'], marker='o', label='Total Efficiency Index')
plt.title('Evolución de la Eficiencia Operativa')
plt.xlabel('Semana (Fecha)')
plt.ylabel('Índice de Eficiencia')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
